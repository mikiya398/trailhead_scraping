```python
def get_tCode_map(sqlite_database_name):
    conn = sqlite3.connect('{}.db'.format(sqlite_database_name))
    c = conn.cursor()
    tCode_dict = defaultdict(str)
    c.execute("SELECT code, name FROM Trailhead order by code")
    records = c.fetchall()
    for record in records:
        tCode = record[0]
        tName = record[1]
        tCode_dict[tName] += tCode      
    conn.close()
    return  tCode_dict
```

```python
def fetch_records_as_dict():
    """
    指定したデータベース・テーブルの内容をディクショナリー型で取得する
    Parameters
    ----------
    sqlite_database_name: String
        取得したいデータベース名

    table_name: String
        取得したいテーブル名

    Returns
    -------
    dict
    """
    conn = sqlite3.connect('example.db')
    conn.row_factory = sqlite3.Row
    c = conn.cursor()
    c.execute("SELECT * FROM Trailhead order by code")
    records = c.fetchall()
    conn.close()

    records_dict = {}
    for record in records:
        records_dict_key = record['name']
        records_dict_value = record['code']
        records_dict[records_dict_key] = records_dict_value
    return  records_dict
```

```python
def csv_to_dict(csv_file):
    data = []
    with open(csv_file, 'r', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            # 必要なキーと値のペアを選択して辞書に追加する
            processed_row = {
                "trailblazerId": row["trailblazerId"],
                "trailheadName": row["trailheadName"],
                "clearedOrder": row["clearedOrder"],
                "pfCode": row["pfCode"],
                "clearedDate": row["clearedDate"]
            }
            data.append(processed_row)
    return data
            
csv_to_dict('/Users/mikiyakitamura/Desktop/desktop/Selenium-Trailhead進捗状況取得/output/2024-04-24-11-02.csv')
```
